```python
import re
import pandas as pd
import numpy as np

# Export Whatsapp Chat without Media


def GetDateTime(m):
    """Getting date and time from the message"""
    try:
        result = re.match(r"\[[^\]]*]", m).group(0)
    except AttributeError:
        result = re.match(r"\[[^\]]*]", m)
    if result:
        return result
    else:
        return None
    
def Getstat(m):
    """Checking if a message has a date and time"""
    try:
        result = re.match(r"\[[^\]]*]", m).group(0)
    except AttributeError:
        return False
    if result:
        return True


def getAuthorAndMessage(m):
    """Extracting Author and Message from text"""
    cleaned_message = m.split(': ')
    message = ' '.join(cleaned_message[1:])
    
    cleaned_author = m.split(' ')
    author = cleaned_author[3]
    return message, author




def parseMessage(m):
    """Parse the message and return a list"""
    message, author = getAuthorAndMessage(m)
    date = GetDateTime(m)
    return [date, author, message]

data = []

with open("_chat.txt", encoding="utf-8") as f:
    f.readline()
    f.readline()
    f.readline()
    counter = 0
    while True:
        line = f.readline()
        if not line:
            break
        line = line.strip()
        if Getstat(line):
            # Recoding the message to get rid of [U+200E]
            str1 = line.encode("ascii", "ignore")
            str2 = str1.decode()
            # Adding Data to the list
            data.append(parseMessage(str2))


df = pd.DataFrame(data, columns=['Date', 'Author', 'Message']) # Initialising a pandas Dataframe.
```

```python
media_messages_df = df[df['Message'] == 'image omitted']
messages_df = df.drop(media_messages_df.index)
messages_df.info()
messages_df['Letter_Count'] = messages_df['Message'].apply(lambda s : len(s))
messages_df['Word_Count'] = messages_df['Message'].apply(lambda s : len(s.split(' ')))
messages_df["MessageCount"]=1
URLPATTERN = r'(https?://\S+)' # Regex for URLs
df['urlcount'] = df.Message.apply(lambda x: re.findall(URLPATTERN, x)).str.len()
```

```python
for i in df.Author.unique():
    # Reading Individual User stats
    req_df= messages_df[messages_df["Author"] == i]
    print("Stats of - {}".format(i))
    print("Messages Sent: {}".format(req_df.shape[0]))
    words_per_message = (np.sum(req_df['Word_Count']))/req_df.shape[0]
    print('Words per message: {}'.format(words_per_message.round()))
    links = sum(req_df["urlcount"])   
    print('Links Sent: {}'.format(links))
    print("\n")
    
    
```
